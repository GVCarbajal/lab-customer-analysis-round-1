```python
import pandas as pd
```

```python
# EXERCISE: Read the three files into python as dataframes

df1 = pd.read_csv('../lab-customer-analysis-round-1/files_for_lab/csv_files/file1.csv')
df1
```

```python
df2 = pd.read_csv('../lab-customer-analysis-round-1/files_for_lab/csv_files/file2.csv')
df2
```

```python
df3 = pd.read_csv('../lab-customer-analysis-round-1/files_for_lab/csv_files/file3.csv')
df3
```

```python
# EXERCISE: Show the DataFrame's shape.

print ('Shape of file1.csv:', df1.shape)
print ('Shape of file2.csv:', df2.shape)
print ('Shape of file3.csv:', df3.shape)
```

```python
# EXERCISE: Standardize header names.

### Standardize column names by transforming them into meaninfull strings, lower cased and with '_' instead of spaces
df1.columns = [df1.columns[col_name].lower().replace(' ','_') for col_name in range(len(df1.columns))]
df2.columns = [df2.columns[col_name].lower().replace(' ','_') for col_name in range(len(df2.columns))]
df3.columns = [df3.columns[col_name].lower().replace(' ','_') for col_name in range(len(df3.columns))]

### Rename 'st' as 'state' for clarity
df1.rename(columns = {'st':'state'}, inplace = True)
df2.rename(columns = {'st':'state'}, inplace = True)
```

```python
# EXERCISE: Rearrange the columns in the dataframe as needed

df2.reindex_like(df1, copy = False)
df3.reindex_like(df1, copy = False)
```

```python
# EXERCISE: Concatenate the three dataframes

df = pd.concat([df1,df2,df3])
```

```python
# EXERCISE: Which columns are numerical?

df_num = df.select_dtypes(include=['number'])
print('Columns containing numerical data:', list(df_num.columns))
```

```python
# EXERCISE: Which columns are categorical?

df_cat = df.select_dtypes(exclude=['number'])
print('Columns containing categorical data:', list(df_cat.columns))
```

```python
# EXERCISE: Understand the meaning of all columns

df.info()
```

```python
# Perform the data cleaning operations mentioned so far in class
# EXERCISE: Delete the column education and the number of open complaints from the dataframe.

df_clean = df.drop(columns=['education', 'number_of_open_complaints'])
```

```python
# EXERCISE: Correct the values in the column customer lifetime value. 
# They are given as a percent, so multiply them by 100 and change dtype to numerical type.

df_clean['customer_lifetime_value'] = df_clean['customer_lifetime_value'].apply(lambda x:float(str(x).replace('%', ''))*100)
```

```python
# EXERCISE: Check for duplicate rows in the data and remove if any.

df_clean.drop_duplicates(inplace=True)
```

```python
# EXERCISE: Filter out the data for customers who have an income of 0 or less.

df_clean = df_clean[df_clean.income > 0]
```

```python
df_clean.to_csv('customer_analysis_round1_clean.csv')
```
